### 2.2 Scraping explanation

Collecting the data was an intensive process in which tens of hours were spent researching web scraping and drafting a Python script. Despite the high intensity of this process, this also makes it very educational. Ultimately, it is not only the delivered script, but mainly the analytical way of thinking that is developed.

In this section I will briefly explain what happened to develop the above script. To start with, the necessary libraries are implemented. Research has shown that Realtor.com has an internal API which is efficient in scraping the data. A connection has been made with this url, headers have been added to circumvent the bot protection so that the scraper is recognized as a device.

The scraping itself consists of several steps (loops). For starters, the Realtor.com website consists of a maximum of 206 pages per search result. This is a total of 42 homes per page (noted as offset: x). The requested query contains information about the page itself. By looking at the queries from different pages it has been possible to discover differences. Firstly, it appears that the page_num increases just like the url and then the offset (the total number of listings) per page increases by 42. A loop has thus been set up to automatically adjust these parameters. The .replace() function is implemented to increase the values ââfor each page from the page_num in the range(1,207) by 1. The offset also had to be changed to avoid scraping only the first page. Per page this was thus increased by 42 (offset += 42). This data is converted to a file by means of json.loads. The second loop iterates over the data in this requested query and places it in the empty prepared list. Because the data was nested (data within data) several parameters were given to the extracting data loop.

In short, the loop goes over a certain number of steps in each iteration. For example, the page_num is increased by 1 and the offset is increased by 42. By sending a new request on the basis of this iteration, it is possible to request the information for each page. This resulted in a document of 8610. With each iteration, new data is added to the compiled list, resulting in a complete data frame at the end of the scrape

# 3. Feature engineering

The basis of feature engineering is to optimize the modeling process of the data and in many cases to prevent the data from being overfitted. In essence, overfitting means that the model is so perfectly tuned to the training and test data that the application does not perform well at new values. To prevent this, it is possible, for example, to generalize certain values by dividing them into groups or categories. In addition, it is possible to create new columns with different content. This leads to possible new insights that are important when developing a statistical model.

### 3.1 Price per square feet

The first feature engineering component that has been applied focuses on calculating price per square foot. In the current literature this is often used to determine the value of a house. For example, a large house may be more expensive than a small house but the relative differences may have other combinations. This can be used to indicate the relative prices of homes.

```python
df5['price_sqft'] = round(df5['price']/df5['sqft'],2)
df5.head()
```

```python
df5 = df5[df5['price_sqft'] < 7000] # Remove all values higher than 1.000.000
```

```python
sns.boxplot(df5['price_sqft']) # Pick right column
plt.title('Year built') # Give title to document
plt.xlabel('Year') # Provide name on x-axis
plt.show() # Present boxplot
```

```python
df_show = df5[['price','price_sqft']]
df_show.head()
```

The above columns demonstrate this with a practical example. Although the houses of rows 1 and 2 are almost identical, the differences in price per square feet are very different. This provides insight into the value of a house in a different way.

### 3.2 Dummy variables

Setting up dummy variables is important when building a machine learning model because it converts categories into quantitative (binary) values. When applying dummies it is necessary to take into account the fact that one column should not be used because this results in an error, the principle of k-1.

```python
df6 = pd.get_dummies((df5['house_type'])) # Get dummies of selected column

merged = pd.concat([df5, df6], axis=1, join='inner')
merged.head()
```

### 3.3 Postal codes

Postcodes is a measurement for geographically distinguishing homes from each other. Although postcodes are shown as ratio values, they are categorical values. Homes with the same zip code are most likely located close to each other. Because in a state like California there are more than 1000 zip codes and not every zip code has enough or at all a home for sale, so it was decided to group this column.

```python
test = df5.groupby('postal_code')['id'].nunique() # count unique values
test
```

```python
bins= [89000,91000,92000,93000,94000,95000,96000] # make bins
labels = ['89k-91k','91k-92k','92k-93k','93k-94k','94k-95k','95k-96k'] # make label names
merged['postal'] = pd.cut(merged['postal_code'], bins=bins, labels=labels, right=False) # create new column with groups
merged.head() # show new column
```

Since postcodes are now categorized in this way, it is again important to set them up as dummy variables to ensure that they can be used in the statistical model.

```python
df7 = pd.get_dummies((merged['postal'])) # Get dummies of selected column

merged = pd.concat([merged, df7], axis=1, join='inner')
merged.head()
```

### 4.1 Distribution (normality)

Basically, a normal distribution is about the symmetry in the variance of the input data. A normal distribution has a perfect composition of information in which the mode, median and mean are all exactly the same. Although this often does not appear in perfect form in reality, plotting this data does provide a certain insight into how the data is distributed. This is important because a poorly distributed data set may say something about the quality of the data used.

#### 4.1.1 Price per square feet

To start with the price per square feet. The plotted diagram initially shows a relatively normally distributed variance. The presence of outliers clearly emerges here, just like in the boxplots. It can therefore be argued that this may not be a normal distribution. On the one hand this is true but the majority shows a clear distributed variance. On the other hand, this division makes sense with regard to the house price itself. In reality, there are more houses with a lower market value than with a higher market value. This is also in line with the unequal distribution of wealth that exists on a global scale, and therefore also for California.

```python
import seaborn as sns # import seaborn library
sns.displot(merged['price_sqft'], kde=False) # show filtered dataset in table and histogram
```

In addition to a subjective view of the dataset, it is also possible to calculate certain characteristics. As stated earlier, the median, mode and mean all say something about the distribution and therefore also about the skewness. In addition to the visual aspect in the graph where a certain degree of skewness can already be detected, which is of course in line with the aforementioned theory. This subjective perception can be reinforced by the actual calculation of it.

```python
mean = merged['price_sqft'].mean() # calculate mean
median = merged['price_sqft'].median() # calculate median
std = merged['price_sqft'].std() # calculate standarddeviation
```

```python
skew = 3 * (mean - median) / std # calculate skewness
round(skew,2)
```

The data shows a result of 0.76 and therefore an extremely strong degree of skewness in the dataset. This again confirms the expectation and observations.

As with the price per square feet, it is again possible to calculate the degree of skewness in this area.

```python
skew = 3 * (mean - median) / std # calculate skewness
round(skew,2)
```

The result shows a skewness of 0.7 which is indeed lower than price per square feet but still shows a significant deviation in symmetry.

# 5. Predictive model

In this piece, all the foregoing material will be elaborated in the design of a predictive model. It will be argued why the specific tool (regression, nearest neighbor, e.g.), parameters and variables were chosen. Various tests were carried out in order to draw up the best possible model. Although the cleanup and exploration process has been carefully carried out, there are a number of things that may have a negative effect on the model. Firstly, it has emerged that there is multicollinearity. The correlations that have been demonstrated have also turned out to be lower than initially expected. This will have a negative effect on the accuracy of the model. Then there is also the effect of the outliers. The great amount of outliers also have a negative effect on the model but I think it is important to make an robust model without large data bias.

A regression model was chosen to draw up this model because it concerns quantitative values and not cluster values. To begin with, multivariate regression was applied. In contrast to linear regression, this method allows the use of multiple variables. A regression analysis model was chosen because it is an accessible and reliable way to gain deep statistical insights into the relationships within the model. It offers the possibility to predict the quantitative dependent variables on the basis of various independent variables. For this model I also tried to put in better parameters to optimize the results but in all cases the standard values of the model itself performed the best. 

### 5.1 Multivariate regression (I)

#### 5.1.1 Implementing model

```python
from sklearn.linear_model import LinearRegression # import linear regression
from sklearn.model_selection import train_test_split # import train_test_split
from sklearn.metrics import mean_squared_error # import mean_squared_error
from sklearn.metrics import r2_score # import r2_score
```

For the first attempt at the model, the variables that contain a relatively strong correlation with the dependent variable were chosen. In order to perfect the model, these values have been adjusted in the following iterations. In this version this is the parts: beds, baths, square feet, stories and year built with dependent variable price instead of price per square feet. This choice was made because the correlation between variables and the price per square feet is so low that this could lead to a very inaccurate model in advance.

```python
Y = merged[['price']] # declare dependent variables
X = merged[['beds','baths','sqft','stories','year_built']] # declare independent variables
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=101) # split data up (30%)
print(X_train.shape) # show shape of data
print(X_test.shape) # show shape of data
print(y_train.shape) # show shape of data
print(y_test.shape) # show shape of data
```

```python
model = LinearRegression() # create new model
model.fit(X_train,y_train) # fit model on training data
```

```python
print(model.intercept_) # print model intercept
```

```python
predictions = model.predict(X_test) # make predictions
predictions
```

```python
sns.regplot(y_test,predictions) # plot regplot to show results
```

#### 5.1.2 Evaluating model

```python
import statsmodels.api as sm
X_train_Sm= sm.add_constant(X_train)
X_train_Sm= sm.add_constant(X_train)
ls=sm.OLS(y_train,X_train_Sm).fit()
print(ls.summary())
```

The score of the current model is very weakly demonstrated by a relatively low R-squared value of 0.474 (47.7%). The adjusted R-squared value is almost the same which is of course explained by the fact that relatively few values have been used (Adjusted R-squared is the "penalty" of adding too many variables to compensate for the possible results ). This weak result shows that the data or the chosen variables do not lead to a model that can be applied in reality.

```python
from sklearn.metrics import mean_squared_error # import mean_squared_error
import math

mse = mean_squared_error(y_test, predictions)

rmse = math.sqrt(mse)
rmse
```

The fact that the model performs poorly is of course also reflected in the RMSE. Because this project involves house prices and the numbers are therefore large, it makes sense to have a higher RSME. In this case, the RSME is close to 900,000. Because average houses cost about as much or less it shows the accuracy is low. This can also be explained by the amount of outliers to which the RSME is sensitive. Similarly, there are 20 million dollar houses and 200,000 dollar houses, which shows that the variance is quite large.

#### 5.1.3 Boosting score

Despite this rather negative result, there are still opportunities to improve this. Before this is done on the basis of adjusted variables, it has emerged from the existing literature that the model can also be improved on the basis of a gradient booster regression (Masui, 2022). This prediction model is applied to generally relatively weak models. What the algorithm does is that it creates weak models and combines them to get a better performing total. Because of this it works better for things like outiers and missing values which is something that is prevalent in the data used. By implementing this technique it is possible to get a higher score on the model.

```python
from sklearn import ensemble # import library
clf = ensemble.GradientBoostingRegressor(n_estimators = 500, max_depth = 5, min_samples_split = 2,
         learning_rate = 0.1, loss = 'ls') # build regression model
```

```python
clf.fit(X_train, y_train) # make model fit
```

```python
clf.score(X_test,y_test) # test score
```

The result shows a value of about 52.4 percent because it is a relatively strong improvement from the previously observed value of 47.4 percent. Although this increase is positive, it is still very low. Mainly because the science of computers and economics sets the standard for an acceptable accuracy at a mininum of around 95%, unlike social sciences where this percentage is significantly lower.

### 5.2 Multivariate regression (II)

#### 5.2.1 Implementing model

In this new model, an attempt was once again made to find variables that yield a better model based on knowledge from the research. It was decided to use the same variables as the first iteration because they still contain a certain level of reliability. It was decided here to also add the postal codes and thus to take the process in spatial context. Because certain zip codes also had a degree of correlation with the price, it was decided to add them. Because these are set up as dummy variables, according to the theory (k-1) one value is not taken into account in order to avoid an error. It must be mentioned that in this notebook is the second iteration but that I have tried more than 30 combinations without any improvements.

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=101)  # split data up (30%)
print(X_train.shape) # show shape of data
print(X_test.shape) # show shape of data
print(y_train.shape) # show shape of data
print(y_test.shape) # show shape of data
```

```python
print(model.intercept_) # print model intercept
```

```python
predictions = model.predict(X_test) # make predictions
predictions
```

```python
sns.regplot(y_test,predictions) # plot regplot to show results
```

#### 5.2.2 Evaluating model

```python
import statsmodels.api as sm
X_train_Sm= sm.add_constant(X_train)
X_train_Sm= sm.add_constant(X_train)
ls=sm.OLS(y_train,X_train_Sm).fit()
print(ls.summary())
```

Even with the implementation of these of the new variables, the outcome still seems relatively low, but there is a clear improvement. For example, the R-squared is now at 0.521 (52.1%) and the adjusted R-squared is now at 0.520 (52.0%). Despite this low standard it is possible to get to a minimum of about 60% by boosting.

```python
mse = mean_squared_error(y_test, predictions)

rmse = math.sqrt(mse)
rmse
```

Also in this almost equivalent model, the extremely high RSME shows that the practical implementation is low, partly due to the large number of outliers.

### 5.3 K-nearest neighbors (KNN)

A final attempt to improve this is by using the nearest neighbor (KNN) method. The reason for doing this is because this model, like the regression, is relatively easy to implement and can handle non-linearity relatively well. The problem this model does have is that, in contrast to multivariate regression, it is less interpretable. For this model, the best values that emerged from the regression test were chosen.

#### 5.3.1 Implementing model

In the case of nearest neighbor, the variables that emerged best from the correlation and multivariate regression tests were chosen. These are the variables with both the strong correlations from the first iteration and the added postal codes.

```python
from sklearn.preprocessing import normalize # import normalize function

X = merged[['beds','baths','sqft','stories','lot_sqft','year_built','89k-91k','91k-92k','92k-93k','93k-94k','94k-95k']]
X = normalize(X) # normalize data
y = merged['price'] # create y variable
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1) # split the data (30%)
```
